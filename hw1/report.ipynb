## 1 项目说明

本项目的文件结构如下：
- `data/FashionMNIST/`：存储了FashionMNIST数据集的训练和测试数据。
- `models/`：保存了训练好的模型文件。
- `train.py`：用于训练模型的脚本。
- `test.py`：用于测试模型的脚本。
- `model.py`：定义了神经网络模型的结构。
- `utils.py`：包含了参数封装，模型封装，自动微分，优化器，以及其他辅助函数的实现。
- `report.ipynb`：用于记录实验过程和结果的Jupyter Notebook。

我们使用最好的参数，训练一个更大的模型
{'LR': 0.0001, 'hiddenlayer': 256, 'nonlin': 'relu', 'weightdecay': 1e-05}

````python
from train import train, draw_curve
from model import MLP
from utils import load_mnist
import numpy as np
 
# Set the parameters for training
params = {'LR': 0.0001, 'hiddenlayer': 256, 'nonlin': 'relu', 'weightdecay': 1e-05}
 
# Load the dataset
img, label = load_mnist("data/FashionMNIST", kind="train")
total_n = len(img)
shuffled_indices = np.random.permutation(total_n)
train_n = int(total_n * 0.9)
train_img = img[shuffled_indices[:train_n]]
train_label = label[shuffled_indices[:train_n]]
valid_img = img[shuffled_indices[train_n:]]
valid_label = label[shuffled_indices[train_n:]]
 
# Initialize the model
model = MLP(hidden1=params['hiddenlayer'], hidden2=params['hiddenlayer'], nonlin=params['nonlin'])
 
# Train the model
train_loss, valid_loss, train_acc, valid_acc, model_snapshots = train(
    model, train_img, train_label, valid_img, valid_label,
    batch_size=64, learning_rate=params['LR'], weight_decay=params['weightdecay'], epoches=20
    )
 
# Visualize the training process
draw_curve(train_loss, valid_loss, train_acc, valid_acc)
````

## 3 测试

````python
from test_model import testmodel
from model import MLP
from utils import load_mnist, load_model
test_img, test_label = load_mnist("data/FashionMNIST", kind="t10k")
model = MLP()
state_dict = load_model("models/model.pth")
model.load(state_dict)

testmodel(model, test_img, test_label)
````

````python
import matplotlib.pyplot as plt

weights = model.fc1.weight.data  # shape: (hidden1, 784)

num_neurons = 36
fig, axes = plt.subplots(6, 6, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    if i < num_neurons:
        weight_image = weights[:,i].reshape(28, 28)  # 重塑为 28x28
        ax.imshow(weight_image, cmap='gray')
        ax.set_title(f'Neuron {i+1}')
        ax.axis('off')
plt.tight_layout()
plt.show()
````

分析：部分图像有一副的轮廓，其中25,26两张图长得像鞋子，说明模型学习到了物体的轮廓。

### 错例分析

````python
from utils import Value, accuracy
import numpy as np
import matplotlib.pyplot as plt

input_val = Value(test_img)
y_pred = model.forward(input_val)
pred_labels = np.argmax(y_pred.data, axis=1)

misclassified_indices = np.where(pred_labels != test_label)[0]

num_samples = 9
selected_indices = np.random.choice(misclassified_indices, num_samples, replace=False)

# 可视化分类错误的样本
CLASS = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
for i, ax in enumerate(axes.flat):
    idx = selected_indices[i]
    ax.imshow(test_img[idx].reshape(28, 28), cmap='gray')
    ax.set_title(f'True: {CLASS[test_label[idx]]}, Pred: {CLASS[pred_labels[idx]]}')
    ax.axis('off')
plt.tight_layout()
plt.show()
````

从得到的错例中可以看出，部分图片不够清晰，导致模型出现了错判，尤其是对于特征相近的类型。
